```python
with open("everybody_codes_e2025_q07_p1.txt","r") as f:
    lines = f.readlines()

names = lines[0].strip().split(",")
rules = lines[2:]
```

```python
rules_parsed = {}
for rule in rules:
    rule = rule.strip()
    parts = rule.split(" > ")
    key = parts[0]
    values = parts[1].split(",")
    rules_parsed[key] = values
```

```python
for name in names:
    name_possible = True
    for i, letter in enumerate(name):
        if i == len(name)-1: # Skip last character in name
            break
        next_letter = name[i+1]
        allowed_letters = rules_parsed[letter]
        if not next_letter in allowed_letters:
            name_possible = False
            break
    print(f"The name '{name}' can {'' if name_possible else 'not '}be constructed with the provided ruleset")
```
