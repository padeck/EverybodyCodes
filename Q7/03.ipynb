```python
with open("everybody_codes_e2025_q07_p3.txt","r") as f:
    lines = f.readlines()

prefixes = lines[0].strip().split(",")
rules = lines[2:]
```

```python
rules_parsed = {}
for rule in rules:
    rule = rule.strip()
    parts = rule.split(" > ")
    key = parts[0]
    values = parts[1].split(",")
    rules_parsed[key] = values
```

```python
def check_prefix(name):
    for x,y in zip(name[:-1], name[1:]):
        res = rules_parsed.get(x)
        if not (res and y in res):
            return False
    return True



def recursion(names, min_letters, max_letters, ):
    new_names = set()
    for name in names:
        if len(name)+1 > max_letters:
            continue
        last_letter = name[-1]
        new_candidates = rules_parsed.get(last_letter)
        if new_candidates:
            for candidate in new_candidates:
                if not name+candidate in names:
                    new_names.add(name+candidate)
    
    if not new_names:
        names = set([name for name in names if len(name)>=min_letters])
        return names
    return recursion(names|new_names, min_letters, max_letters)

prefixes = set([prefix for prefix in prefixes if check_prefix(prefix)])
unique_names = recursion(prefixes, 7, 11)
```
