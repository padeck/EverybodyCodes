```python
with open("everybody_codes_e2025_q07_p3.txt") as f:
    data = f.read().split("\n\n")
    prefixes = data[0].split(",")
    lines = data[1].splitlines()
```

```python
rules_parsed = {}
for rule in rules:
    rule = rule.strip()
    parts = rule.split(" > ")
    key = parts[0]
    values = parts[1].split(",")
    rules_parsed[key] = values
```

```python
def check_prefix(name):
    for x,y in zip(name[:-1], name[1:]):
        res = rules_parsed.get(x)
        if not (res and y in res):
            return False
    return True
```

```python
total = set()
def recursion(current_name, min_letters, max_letters):
    if min_letters <= len(current_name) <= max_letters:
        total.add(current_name)
    if len(current_name) > max_letters:
        return
    
    possible_options = rules_parsed.get(current_name[-1])
    
    if possible_options:
        for option in possible_options:
            recursion(current_name+option, min_letters, max_letters)

```

```python
for prefix in prefixes:
    if check_prefix(prefix):
        recursion(prefix, 7, 11)
```
