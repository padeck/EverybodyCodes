```python
with open("everybody_codes_e2025_q07_p2.txt","r") as f:
    lines = f.readlines()

names = lines[0].strip().split(",")
rules = lines[2:]
```

```python
rules_parsed = {}
for rule in rules:
    rule = rule.strip()
    parts = rule.split(" > ")
    key = parts[0]
    values = parts[1].split(",")
    rules_parsed[key] = values
```

```python
indices_sum = 0
for i, name in enumerate(names):
    name_possible = True
    for j, letter in enumerate(name):
        if j == len(name)-1: # Skip last character in name
            break
        next_letter = name[j+1]
        allowed_letters = rules_parsed[letter]
        if not next_letter in allowed_letters:
            name_possible = False
            break
    indices_sum += (i + 1) if name_possible else 0
print(indices_sum)
```
