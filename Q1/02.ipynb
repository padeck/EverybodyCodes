```python
names = ["Vyrdax","Drakzyph","Fyrryn","Elarzris"]

instructions = ["R3","L2","R3","L1"]
```

```python
with open("everybody_codes_e2025_q01_p2.txt", "r") as file:
    lines= file.readlines()

names = lines[0].split(',')
instructions = lines[2].split(',')
```

```python
def move(current_position:int, instruction: str):
        parsed = parse_instruction(instruction)
        if parsed is None:
            print("Invalid instruction:", instruction)
            return current_position
        direction, step_amount = parsed
        if direction == "L":
            current_position = (current_position - step_amount) % len(names)
            return current_position
        if direction == "R":
            current_position = (current_position + step_amount) % len(names)
            return current_position
        else:
            print("Error")
            return current_position




def parse_instruction(instruction: str):
    direction = instruction[0]
    amount = int(instruction[1:])
    return direction, amount
```

```python
current_position = 0

for instruction in instructions:
    current_position = move(current_position, instruction)

print(f'The name for the final step is: {names[current_position]}')
```
