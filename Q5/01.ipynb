```python
from SpineSegment import SpineSegment
```

```python
with open("everybody_codes_e2025_q05_p1.txt", "r") as f:
    line = f.readline()
```

```python
nums = line.split(":")[1]
nums = nums.split(",")
nums = [int(x) for x in nums]
```

```python
spine_segs: list[SpineSegment] = []

def handle_init(obj):
    spine_segs.append(obj)
```

```python
for n in nums:
    added = False

    # First check if the number fits in an existing spine segment
    for seg in spine_segs:
        if seg.is_complete: # Skip completed segments
            continue

        added = False
        added = seg.add_element(n)
        if added: 
            break
    if added:
        continue
    
    # Otherwise create a new spine segment
    SpineSegment(n, handle_init)
```

```python
for seg in spine_segs:
    print(seg)
```

```python
quality_string = ""
for seg in spine_segs:
    quality_string += str(seg.middle)
quality_string = int(quality_string)
quality_string
```
