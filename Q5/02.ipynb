```python
from SpineSegment import SpineSegment
```

```python
with open("everybody_codes_e2025_q05_p2.txt", "r") as f:
    lines = f.readlines()
    spine_segs_zippped = []
    for line in lines:
        id = line.split(":")[0]
        nums = line.split(":")[1]
        nums = nums.split(",")
        nums = [int(x) for x in nums]
        spine_segs_zippped.append((id,nums))
```

```python
def determine_quality(spine_segs):
    quality_string = ""
    for seg in spine_segs:
        quality_string += str(seg.middle)
    quality = int(quality_string)
    return quality
```

```python
qualities: list[int] = []
for id, nums in spine_segs_zippped:
    spine_segs: list[SpineSegment] = []

    def handle_init(obj):
        spine_segs.append(obj)
        
    for n in nums:
        added = False

        # First check if the number fits in an existing spine segment
        for seg in spine_segs:
            if seg.is_complete: # Skip completed segments
                continue

            added = False
            added = seg.add_element(n)
            if added: 
                break
        if added:
            continue
        
        # Otherwise create a new spine segment
        SpineSegment(n, handle_init)
    quality = determine_quality(spine_segs)
    qualities.append(quality)
```

```python
qualities.sort()
highest = max(qualities)
lowest = min(qualities)
print(f'The quality difference is {highest-lowest}')
```
