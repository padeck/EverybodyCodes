```python
from SpineSegment import SpineSegment
from Spine import Spine
```

```python
with open("everybody_codes_e2025_q05_p3.txt", "r") as f:
    lines = f.readlines()
    spine_segs_zippped = []
    for line in lines:
        id:int = int(line.split(":")[0])
        nums = line.split(":")[1]
        nums = nums.split(",")
        nums = [int(x) for x in nums]
        spine_segs_zippped.append((id,nums))
```

```python
spines: list[Spine] = []
for id, nums in spine_segs_zippped:
    spine_segs: list[SpineSegment] = []

    def handle_init(obj):
        spine_segs.append(obj)
        
    for n in nums:
        added = False

        # First check if the number fits in an existing spine segment
        for seg in spine_segs:
            if seg.is_complete: # Skip completed segments
                continue

            added = False
            added = seg.add_element(n)
            if added: 
                break
        if added:
            continue
        
        # Otherwise create a new spine segment
        SpineSegment(n, handle_init)
    spine = Spine(id, spine_segs)
    spines.append(spine)
```

```python
from functools import cmp_to_key

def compare(a: Spine, b: Spine) -> int:
    if a.quality < b.quality:
        return -1 # a before b
    if a.quality > b.quality:
        return 1 # b before a
    if a.quality == b.quality:
        return a.compare_segments(b)
    
    return 0

sorted_list = sorted(spines, key=cmp_to_key(compare))
```

```python
checksum = 0

for i in range(len(sorted_list)):
    checksum += i*sorted_list[i].id

checksum
```
