```python
from tqdm import tqdm
```

```python
with open("everybody_codes_e2025_q08_p2.txt") as f:
    edge_notes = [int(i) for i in f.readline().split(",")]
```

```python
def get_edges(edge_notes):
    edges = []
    for i in range(len(edge_notes)-1):
        edges.append((edge_notes[i], edge_notes[i+1]))
    return edges

edges = get_edges(edge_notes)
```

```python
def check_intersection(edge_1, edge_2) -> bool:
    min_1 = min(edge_1)
    min_2 = min(edge_2)
    max_1 = max(edge_1)
    max_2 = max(edge_2)

    # If there are any shared nodes, no overlap exists
    for e_1 in edge_1:
        for e_2 in edge_2:
            if e_1 == e_2:
                return False

    # Check other cases
    # First Check this case:
    # |-----|
    #   |-----|
    if min_1 < min_2 and max_1 < max_2 and min_2 < max_1:
        return True

    # Now check the other case:
    #   |----|
    # |----|
    if min_2 < min_1 and max_2 < max_1 and min_1 < max_2:
        return True

    return False
```

```python
visited_edges = []
total_intersections = 0
for edge in tqdm(edges):
    for visited in visited_edges:
        if check_intersection(edge, visited):
            total_intersections += 1
    visited_edges.append(edge)
```
