```python
with open("everybody_codes_e2025_q06_p3.txt", "r") as f:
    line = f.readline()
```

```python
def determine_boundary(index, char_string_length, reach):
    # Lower End
    lower_end = index - reach
    if lower_end < 0:
        lower_end = 0
    
    # Upper End
    upper_end = index + reach
    if upper_end > char_string_length:
        upper_end = char_string_length
    
    return lower_end, upper_end

def determine_mentor_combination_amount(char_string, reach, pattern_repetition_amount):
    total = 0
    char_string = char_string * pattern_repetition_amount

    for i in range(len(char_string)):
        c = char_string[i]
        if c.islower():
            lower, upper = determine_boundary(i, len(char_string), reach)
            partial_string = char_string[lower : upper + 1]
            mentors = partial_string.count(c.upper()) # Count the mentors for the partial string and add them to total
            total += mentors
    return total
```

```python
determine_mentor_combination_amount(line, 1000, 1000)
```
