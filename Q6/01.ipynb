## Q1

```python
with open("everybody_codes_e2025_q06_p1.txt", "r") as f:
    line = f.readline()

line
```

```python
search_chars = ["a"]
possibilities = 0

for s in search_chars:

    cap = s.upper()
    mentors = 0
    novices = 0

    for char in line:
        if char == cap: # Mentor found
            mentors += 1
        if char == s: # Novice found
            possibilities += mentors

possibilities
```
