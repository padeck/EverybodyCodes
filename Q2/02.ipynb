```python
from ComplexNumber import ComplexNumber
from tqdm import tqdm
import ast
```

```python
with open("everybody_codes_e2025_q02_p2.txt", "r") as f:
    line = f.read().strip()  # Read the line

# Split at '=' to separate the variable name and value
var_name, var_value = line.split("=")
val1 = var_name.strip()
A = ast.literal_eval(var_value)
A = ComplexNumber(A[0],A[1])
```

```python
def check_engraved(coord:ComplexNumber) -> bool:
    result = ComplexNumber(0,0)
    divCN = ComplexNumber(100000,100000)
    for _ in range(100):
        result = result * result
        try:
            result = result / divCN
        except OverflowError:
            return False
        result = result + coord
    if (abs(result.x) > 1e6) or (abs(result.y) > 1e6):
        return False
    return True

def calculate_engravements(precision: int=10) -> int:
    counter = 0
    for i in tqdm(range(0,1001,precision), desc="Rows"):
        for j in range(0,1001,precision):
            coord = ComplexNumber(A.x+j,A.y+i)
            if check_engraved(coord):
                counter += 1
    return counter
```

```python
calculate_engravements(precision=1)
```
