```python
from math import prod, floor
```

```python
with open("everybody_codes_e2025_q04_p3.txt", "r") as f:
    lines = f.readlines()  # Read the line
    gears = []
    for line in lines:
        pair = line.split("|")
        if len(pair) == 2:
            gears.append((int(pair[0]), int(pair[1])))
        else:
            gears.append(int(pair[0]))
```

```python
ratios = []
for idx in range(0, len(gears)-1):
    curr = gears[idx]
    nxt = gears[idx + 1]

    # Extract current gear value
    val1 = curr if isinstance(curr, int) else curr[1]
    # Extract next gear value
    val2 = nxt if isinstance(nxt, int) else nxt[0]

    ratios.append(val1 / val2)

product = prod(ratios)
```

```python
print(f'The last gear will turn {floor(100 * product)} turns')
```
