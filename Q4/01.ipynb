```python
with open("everybody_codes_e2025_q04_p1.txt", "r") as f:
    lines = f.readlines()  # Read the line
    gears = []
    for line in lines:
        gears.append(int(line))
```

```python
def calc_product(ratios: list[float]) -> float:
    product = 1
    for ratio in ratios:
        product *= ratio
    return product
```

```python
ratios = []
for idx in range(0, len(gears)):
    if idx > 0: # Skip the first gear
        idx_prev = idx - 1
        ratio = gears[idx_prev] / gears[idx]
        ratios.append(ratio)

prod = calc_product(ratios)
```

```python
print(f'The last gear will complete {prod * 2025} full turns')
```
