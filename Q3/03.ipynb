```python
with open("everybody_codes_e2025_q03_p3.txt", "r") as f:
    line = f.readline()  # Read the line
    crates = [int(x) for x in line.split(",")]
    crates.sort(reverse=True)
```

```python
# Continually remove ordered sets from the original lists. Ordered sets are the optimal way of the described challenge
ctr = 0
while len(crates) > 0:
    ctr += 1
    crate_set = list(set(crates))
    crate_set.sort(reverse=True)
    for crate in crate_set:
        crates.remove(crate)
```
