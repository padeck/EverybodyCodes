```python
with open("everybody_codes_e2025_q03_p2.txt", "r") as f:
    line = f.readline()  # Read the line
    numbers = [int(x) for x in line.split(",")]
```

```python
def calc_crates(crates: list[int]) -> int:
    crates = list(set(crates))
    crates.sort(reverse=True)
    crates = crates[-20:]   # Only get the last 20 elements after sorting
    return sum(crates)
```

```python
calc_crates(numbers)
```
